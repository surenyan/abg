```python
import sympy as sp
x = sp.symbols('x')
f = x**3 -6*x**2 +9*x + 1
df = sp.diff(f, x)
print("f(x) =", f)
print("f'(x) =", df)
roots = sp.solve(df, x)
print("Roots of derivative:", roots)
```

```python
a = 1
b = 5
n = 8
epsilon = 0.5
```

```python
f_num = sp.lambdify(x, f)
x_vals = []
f_vals = []

for k in range(n+1):
    xk = a + k*((b-a)/n)
    fx = f_num(xk)

    x_vals.append(xk)
    f_vals.append(fx)

    print("k =", k, "xk =", xk, "f(xk) =", fx)

xmin = min(x_vals, key=f_num)
fmin = f_num(xmin)

print("\nxmin=",xmin)
print("f_min=", fmin)
```

```python
x_star = None
for r in roots:
    if r.is_real and a <= r <= b:
        if x_star is None or f_num(r) < f_num(x_star):
            x_star = r

print("x* =", x_star)
```

```python
print("|xmin - x*| =", abs(xmin - x_star))

if abs(xmin - x_star) <= epsilon:
    print("Condition satisfied")
else:
    print("Condition not satisfied")
```

```python
import numpy as np
import plotly.graph_objects as go

x_plot = np.linspace(a, b, 400)
y_plot = f_num(x_plot)

fig = go.Figure()

# Smooth function line
fig.add_trace(go.Scatter(
    x=x_plot, y=y_plot,
    mode='lines', name='f(x)', line=dict(color='blue')
))

# All selection method points (orange)
fig.add_trace(go.Scatter(
    x=x_vals, y=f_vals,
    mode='markers', name='Selection points',
    marker=dict(color='orange', size=8),
    hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'
))

# Minimum point from selection (red)
fig.add_trace(go.Scatter(
    x=[xmin], y=[fmin],
    mode='markers', name='Minimum by selection',
    marker=dict(color='red', size=32, symbol='star'),
    hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'
))

# Actual extremum x* (green)
fig.add_trace(go.Scatter(
    x=[float(x_star)], y=[float(f_num(x_star))],
    mode='markers', name='Actual extremum x*',
    marker=dict(color='green', size=20, symbol='star'),
    hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'
))

fig.update_layout(
    title='Function f(x) and min',
    xaxis_title='x',
    yaxis_title='f(x)',
    hovermode='closest',
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```
