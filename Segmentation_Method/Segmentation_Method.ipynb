```python
import sympy as sp
import numpy as np
import plotly.graph_objects as go

x = sp.symbols('x')
f = x**3 - 6*x**2 + 9*x + 1
f_num = sp.lambdify(x, f)

a, b = 1, 5
epsilon = 0.5
delta = 0.125
```

```python
# Actual minimum via derivative
roots = sp.solve(sp.diff(f, x), x)
x_star = min((r for r in roots if r.is_real and a <= r <= b), key=f_num)
```

```python
# Bisection method
def bisection(a, b, eps, d):
    midpoints = []
    k = 0
    while abs(b - a) > eps:
        x1, x2 = (a + b - d) / 2, (a + b + d) / 2
        print(f"\nIteration {k}: a={a:.4f},  b={b:.4f},  x1={x1:.4f},  f(x1)={f_num(x1):.4f},  x2={x2:.4f},  f(x2)={f_num(x2):.4f}")
        if f_num(x1) <= f_num(x2):
            b = x2
        else:
            a = x1
        mid = (a + b) / 2
        midpoints.append(mid)
        print(f"  new interval: a={a:.4f}, b={b:.4f}, mid={mid:.4f}, |b-a|={abs(b-a):.4f}")
        k += 1
    print(f"\n |b-a| = {abs(b-a):.4f} <= epsilon={eps}")
    print(f"Final: a={a:.4f}, b={b:.4f}, xmin = (a+b)/2 = ({a:.4f}+{b:.4f})/2 = {(a+b)/2:.4f}")
    return (a + b) / 2, midpoints
xmin, midpoints = bisection(a, b, epsilon, delta)
```

```python
print("x* (actual)   =", x_star,  "| f(x*) =", f_num(x_star))
print("xmin (dichot) =", xmin,    "| f(xmin) =", f_num(xmin))
print("|xmin - x*|   =", abs(xmin - float(x_star)))
print("Condition", "satisfied" if abs(xmin - float(x_star)) <= epsilon else "not satisfied")
```

```python
# Plot
x_plot = np.linspace(a - 0.5, b + 0.5, 400)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_plot, y=f_num(x_plot), mode='lines', name='f(x)', line=dict(color='blue')))
fig.add_trace(go.Scatter(x=midpoints, y=[f_num(p) for p in midpoints], mode='markers', name='Iteration midpoints',
    marker=dict(color='orange', size=8), hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'))
fig.add_trace(go.Scatter(x=[xmin], y=[f_num(xmin)], mode='markers', name='Minimum by dichotomy',
    marker=dict(color='red', size=14, symbol='star'), hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'))
fig.add_trace(go.Scatter(x=[float(x_star)], y=[float(f_num(x_star))], mode='markers', name='Actual minimum x*',
    marker=dict(color='green', size=20, symbol='star'), hovertemplate='x = %{x:.4f}<br>f(x) = %{y:.4f}<extra></extra>'))

fig.update_layout(title='Dichotomy Method — f(x) = x³ - 6x² + 9x + 1',
    xaxis_title='x', yaxis_title='f(x)', hovermode='closest', legend=dict(x=0.01, y=0.99))
fig.show()
```
